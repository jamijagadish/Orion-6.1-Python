```python
import matplotlib.pyplot as plt
```

```python
company = ['TCS', 'RPW', 'OLA', 'JPW']
revenue = [123000, 156000, 90000, 105000]
plt.bar(company,revenue)
```

```python
plt.barh(company,revenue, color='r')
```

```python
import numpy as np
comp = np.arange(len(company))

plt.xticks(comp,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(comp-0.3,revenue, width=0.3, label='Revenue')

profit=[93000, 86000, 24000, 55000]
plt.bar(comp+0,profit, width=0.3, label='Profit')

loss = [45000, 34000, 60000, 53000]
plt.bar(comp+0.3, loss, width=0.3, label="Loss")
plt.legend()
```

# Scatter Plot

```python
x = [1, 2, 3, 4, 5]
y = [5, 7, 4, 6, 8]

plt.scatter(x, y, color='green', label='Plot 1')
plt.title("Scatter Plot")
plt.grid()
plt.legend()
```

# Histogram

```python
data = np.random.randn(1000)
plt.hist(data)
plt.title("Histogram")
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=5)
plt.title("Histogram")
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=6, color='maroon', rwidth=0.9)
plt.title("Histogram")
```

```python
data=[93000, 86000, 74000, 95000, 88000, 66000, 76000]
plt.hist(data, bins=[60000, 75000, 90000, 100000], color='purple', rwidth=1)
```

```python
data=[93000, 86000, 74000, 55000, 88000]
plt.hist(data, bins=[60000, 75000, 90000, 100000], color='purple', histtype='step')
```

```python
data1=[93000, 86000, 74000, 55000, 88000]
data2=[95000, 82000, 70000, 59000, 68000]
data3=[93000, 87000, 70000, 72000, 68000]
plt.hist([data1, data2, data3], bins=[50000, 70000, 85000, 100000], color=['green','purple', 'yellow'])
```

```python
data1=[93000, 59000, 86000, 74000, 55000, 88000]
data2=[95000, 82000, 83000, 62000, 59000, 68000]
plt.xlabel("Values")
plt.ylabel("Counts")
plt.hist([data1, data2], bins=[50000, 70000, 90000, 100000], color=['green','yellow'], rwidth=0.9, label=['Data 1', 'Data 2'])
plt.title("Histogram Datas")
plt.legend()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='black', edgecolor='yellow')
plt.title("Histogram")
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', edgecolor='black', orientation='horizontal')
plt.title("Histogram")
plt.savefig('histogram.jpg')
```

# Pi-chart

```python
label = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=label)
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=label, radius=1.2)
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 25, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True)
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 35, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, explode=[0,0.1,0.3,0])
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=60)
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
colors = ['gold', 'aqua', 'maroon', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Programming Language Usage")
plt.savefig('pieChart.jpg')
```
