```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [5,10,20,15,30,35]

plt.plot(x, y, color='y', linestyle='-.', marker='x' , markersize=10, linewidth=4)
plt.title("Simple Line Plot")
```

```python
plt.plot(x, y, color='g', linestyle='-.', marker='o' , linewidth=2)
plt.title("Simple Line Plot")
plt.xlabel("Temperature")
plt.ylabel("Wind Speed")
```

```python
plt.plot( [10, 20, 25, 30], label='Sales 2024', color='b', marker='x')
plt.plot( [15, 18, 22, 28], label='Sales 2025', color='g', marker='x')
plt.plot( [12, 10, 18, 25], label='Sales 2026', color='purple', marker='x')
plt.legend()
```

```python
plt.plot([1,2,3],[3,4,5], label='Line A')
plt.plot([1,2,3],[5,3,1], label='Line B')
plt.title("Example Plot")
plt.legend(loc='best', title='Legend Title', fontsize=10, frameon=True)
```

# Bar graph

```python
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 12]

plt.bar(x, y, color='green')
plt.title("Bar Chart")
plt.xlabel("Name")
plt.ylabel("Values")
```

```python
import numpy as np

company = ['TCS', 'RPW', 'OLA', 'JPW']
revenue = [123000, 156000, 90000, 105000]
comp = np.arange(len(company))
plt.bar(comp,revenue)
```

```python
plt.xticks(comp,company)
plt.bar(comp,revenue)
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(company,revenue, label='Revenue')
plt.legend()
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("Companies Revenue")
plt.bar(company,revenue, label='Revenue')

profit=[93000, 86000, 24000, 55000]
plt.bar(company,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(comp,company)
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.bar(comp-0.2,revenue, width=0.4, label='Revenue')

profit=[93000, 86000, 24000, 55000]
plt.bar(comp+0.2,profit, width=0.4, label='Profit')
plt.legend()
```
