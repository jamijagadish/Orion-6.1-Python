```python
import pandas as pd
```

# Group by

```python
df = pd.DataFrame({
    "City": ['Mumbai','Mumbai','Mumbai', 'Bhubaneswar', 'Bhubaneswar', 'Bhubaneswar', 'Bengaluru','Bengaluru','Bengaluru'],
    "temp": [12,13,19,41,31,40,38,29,40],
    "wind": [5,6,23,5,8,9,1,9,6]})
df
```

```python
gr = df.groupby('City')
gr
```

```python
for city, cityDf in gr:
    print(city)
    print(cityDf)
```

```python
gr.get_group("Bhubaneswar")
```

```python
gr.max()
```

```python
gr.min()
```

```python
gr.mean()
```

```python
gr.describe()
```

```python
gr1 = df.groupby('wind')
```

```python
for speed, group in gr1:
    print(speed)
    print(group)
```

# Concatenation 

```python
weather1 = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19],
    "wind": [5,6,23]})
weather1
```

```python
weather2 = pd.DataFrame({
    "City": ['Chennai', 'Delhi','Hyderabad'],
    "temp": [23,23,29],
    "wind": [5,6,9]})
weather2
```

```python
df = pd.concat([weather1, weather2])
df
```

```python
df = pd.concat([weather1, weather2], ignore_index=True)
df
```

```python
df = pd.concat([weather1, weather2], keys=["DF1", "DF2"])
df
```

```python
df.loc['DF2']
```

```python
df = pd.concat([weather1, weather2], axis= 1)
df
```

```python
df = pd.concat([weather1, weather2], axis= 0)
df
```

```python
df1 = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19],
    "wind": [5,6,23]}, index = [1,2,3])
df1
```

```python
df2 = pd.DataFrame({
    "City": ['Mumbai', 'Bhubaneswar','Bengaluru'],
    "temp": [12,13,19],
    "wind": [5,6,23]}, index = [2,1,3])
df2
```

```python
df = pd.concat([df1,df2], axis=1)
df
```
