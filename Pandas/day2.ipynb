```python
def convert_gender(cell):
    if cell == 'Female': 
        return 'Women'
    elif cell == 'Male':
        return 'Men'

def convert_country(cell):
    if cell == 'United States':
        return 'USA'

df3 = pd.read_excel('file_example.xls', sheet_name='Data', converters={'Gender': convert_gender, 'Country': convert_country})
df3
```
